```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

import numpy as np
from sklearn.ensemble import RandomForestRegressor

import pandas as pd

N = int(input())
tmax = list()
tmin = list()
missing = list()
headers = input().strip().split()
for i in range(N):
    mtuple = input().strip().split()
    tmax.append(mtuple[2])
    tmin.append(mtuple[3])
    # find the missing entries.
    if(str(mtuple[2] + mtuple[3]).__contains__("Missing")):
        missing.append(i)

# do data frame for ease of to num handeling
train = pd.DataFrame({"tmax" : tmax, "tmin" : tmin}, dtype = float)
train["tmin"] = pd.to_numeric(train["tmin"], errors='coerce')
train["tmax"] = pd.to_numeric(train["tmax"], errors='coerce')

# at what index do we need to predict=
tmin_miss = train["tmin"].isna()
tmax_miss = train["tmax"].isna()

# prepare for training
X1_train = np.array(train[~tmin_miss].index)
X1_train = X1_train.reshape(X1_train.shape[0],1)
y1_train = np.array(train.tmin[~tmin_miss],dtype="float")

X1_test = np.array(train[tmin_miss].index)
X1_test = X1_test.reshape(X1_test.shape[0],1)

X2_train = np.array(train[~tmax_miss].index)
X2_train = X2_train.reshape(X2_train.shape[0],1)
y2_train = np.array(train.tmax[~tmax_miss],dtype="float")

X2_test = np.array(train[tmin_miss].index)
X2_test = X2_test.reshape(X2_test.shape[0],1)

# init models
model_tmin = RandomForestRegressor(random_state=0)
model_tmax = RandomForestRegressor(random_state=0)

# train models
model_tmin.fit(X1_train,y1_train)
model_tmax.fit(X2_train,y2_train)

# predict
predict_tmin = model_tmin.predict(X1_test)
predict_tmax = model_tmax.predict(X2_test)

j_min = 0
j_max = 0
for ind in missing:
    if(np.isin(ind, train.tmin[tmin_miss].index)):
        print(predict_tmin[j_min])
        j_min += 1
    
    if(np.isin(ind, train.tmax[tmax_miss].index)):
        print(predict_tmax[j_max])
        j_max += 1

```
