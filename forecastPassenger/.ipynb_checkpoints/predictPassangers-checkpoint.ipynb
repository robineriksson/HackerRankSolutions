```python
T = 12

def armaforcast(data, P=2, D=1, Q=1):
    model = ARIMA(data, order=(P,D,Q))
    model_fit = model.fit(disp=False)
    yhat = model_fit.forecast()[0]
    return yhat
```

```python

def olsforcast(data):
    X = range(1,len(data))
    X = sm.add_constant(X)
    model = sm.OLS(data,X)
    model_fit = model.fit(disp=False)
    yhat = model_fit.predict()[-1]
    return yhat
```

```python
seendata = [x for x in train]
```

```python
olforcast(seendata)
```

```python

predicted = list()
for i in range(T):
    yhat = armaforcast(seendata, P=2, D=1,Q=1)
    yhat2 = olsforcast(seendata)
    predicted.append(yhat)
    print(yhat, yhat2)
    seendata.append(yhat)
```
